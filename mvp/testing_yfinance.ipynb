```python
# get historical market data
hist = msft.history(period="max")
hist = hist[['Open', 'High', 'Low', 'Close']]
hist
```

```python
today = hist[len(hist) - 1:]
today
```

```python
today = today.assign(Name = msft.info['shortName'])
```

```python
today
```

```python
today_json = today.to_json(orient='records')
print(type(today_json))
parsed = json.loads(today_json)
parsed[0].update({'test':123})
#json.dumps(parsed)
print(parsed)
```
